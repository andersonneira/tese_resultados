```python
from mpl_toolkits.mplot3d.proj3d import proj_transform
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib.text import Annotation

class Annotation3D(Annotation):

    def __init__(self, text, xyz, *args, **kwargs):
        super().__init__(text, xy=(0, 0), *args, **kwargs)
        self._xyz = xyz

    def draw(self, renderer):
        x2, y2, z2 = proj_transform(*self._xyz, self.axes.M)
        self.xy = (x2, y2)
        super().draw(renderer)
        
def _annotate3D(ax, text, xyz, *args, **kwargs):
    '''Add anotation `text` to an `Axes3d` instance.'''

    annotation = Annotation3D(text, xyz, *args, **kwargs)
    ax.add_artist(annotation)

setattr(Axes3D, 'annotate3D', _annotate3D)
```

# AvaliaÃ§Ã£o 3

```python
limit = 5632
init = 3294
count = 0
colunas = ['10_kurt_total_ips_origem', '10_skw_total_ips_destino', '10_coefficient_variation_total_pacotes' ]
resultado_4 = []
for i in range(init,limit - 500):
    novo_limit = i + 500   
    data = early_warning[colunas].copy()[init:novo_limit]
    y_real = early_warning['has_bot'][init:novo_limit]
    
    clustering = BisectingKMeans(n_clusters=2, random_state=0)
    y = clustering.fit_predict(data)
    cdbw = compute_cdbw(data, y)
    if cdbw >= 39.043326:
        print("Executando o intervalo: ", i, novo_limit)
        acc = accuracy_score(y_real, y)
        if acc < 0.5:
           y = [0 if i == 1 else 1 for i in clustering.labels_]
        print(confusion_matrix(y_real, y)) 
        print(classification_report(y_real, y, digits=6))
        count = count + 1
        
print("o sistema gerou ", count, ' alertas')
   
```
